```python
class BankAccount():
    def __init__(self,account_number,account_holder,balance):
        self.__account_number=account_number
        self.__account_holder=account_holder
        self.__balance=balance
        self.__transactions=[]
    
    def deposit(self,amount):#Adds the specified amount to the account balance. 
        self.__balance=self.__balance+amount
        self.__transactions.append(f'${amount} is added to account\n new Balance is {self.__balance}')
        return f'${amount} is added to account'
    
    def withdraw(self,amount):#Deducts the specified amount from the account balance if sufficient funds are available.
        if amount > self.__balance:
            print("insufficient Balance")
            return False
        elif amount <= self.__balance:
            self.__balance=self.__balance-amount
            self.__transactions.append(f'${amount} is Subracted  from account\n new Balance is {self.__balance}')
            return f'${amount} is subracted from account'
        elif amount<0:
            print("please enter a positive number")
            return False
        else:
            print("Something Went wrong")
            return False
    
    def check_balance(self):#Returns the current account balance.
        return f"Youre current Balance is ${self.__balance}"
    
    def add_transaction(self,description):#Adds a description of a transaction to the transactionhistory. 
        
        self.__transactions.append(f'{description}') 
     
    def print_statement(self):# Prints a detailed statement of all transactions
        
        for i in range(self.__transactions):
            print(i)
                
            
```

```python
class Bank(BankAccount):
    def __init__(self,account_number,account_holder,balance,accounts={}):
        super().__init__(account_number,account_holder,balance)
        self.__accounts=dict(accounts)

    def open_account(self,account_holder):
        self.__accounts[self.__account_number]= {
            
            'name' : account_holder,
            'balance':self.__balance,
            'transactions':self.__transactions
            
        }
    def get_account(self,account_number): #Retrieves an account object using its account number. 
        print(self.__accounts[account_number])
    
    def transfer(self,sender_account_number, receiver_account_number, amount): #Transfersmoney between two accounts. 
        self.__accounts[receiver_account_number]['balance']=self.__accounts[receiver_account_number]['balance']+amount
        self.__accounts[receiver_account_number]['transactions'].append(f"{amount} is added")
        self.__accounts[sender_account_number]['balance']=self.__accounts[sender_account_number]['balance']-amount
        self.__accounts[sender_account_number]['transactions'].append(f"{amount} is subracted")

    def admin_check_total_deposit(self):# Returns the total balance of all accounts in the bank.
        for i in self.__accounts:
            print(f'the total doposites in {self.__accounts[i]["name"]} is {self.__accounts[i]["balance"]}')
    
    def admin_check_total_accounts():# Returns the total number of accounts in the bank.
        for i in self.__accounts:
            d=0
            d=d+1
        print(f"the total number of account are {d}")
```
